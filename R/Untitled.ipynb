```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
gapminder_r_url = "https://raw.githubusercontent.com/jennybc/gapminder/master/data-raw/08_gap-every-five-years.tsv"
gapminder = pd.read_csv(gapminder_r_url, sep="\t")
gapminder.shape
```

```python
gapminder.describe()
```

```python
gapminder.query("country=='Korea, Rep.' & year==2007")
```

```python
gapminder[["pop", "gdpPercap"]].head()
```

```python
gapminder.\
    assign(total_gdp = lambda x: (x['pop'] * x['gdpPercap'])).\
    assign(le_gdp_ratio = lambda x: (x['lifeExp'] / x['gdpPercap'])).\
    assign(lgrk = lambda x: x['le_gdp_ratio'] * 100).\
    head()
```

```python
gapminder.aggregate(['mean', 'median'])
```

```python
np.random.seed(12345)
gapminder.sample(n=10)
```

```python
gapminder.country.unique(), gapminder.year.unique()
```

```python
gapminder.\
    query("country=='Korea, Rep.'").\
    groupby('country').\
    agg({'lifeExp':'mean'})
```
